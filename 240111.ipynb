```python
import numpy as np
x=np.arange(1,11)
y=np.random.randint(1,30,10)
plt.scatter(x,y,c='k',marker='^')#• scatter( x값, y값, c=‘색 문자’ 혹은 ‘색깔 문자열’, marker=‘마커기호’, label=‘그래프이름’ )
plt.show()
```

```python
# bar(x값, y값, color=‘색 문자’ 혹은 ‘색깔 문자열’) ->세로 막대 그래프
# barh(x값, y값, color=‘색 문자‘ 혹은 ‘색깔 문자열‘) ->가로 막대 그래프
x = ['A','B','C']
y = [100, 150, 120]
plt.bar(x,y)
plt.show()
```

```python
x = ['A','B','C']
y = [100, 150, 120]
plt.barh(x,y)
plt.show()
```

```python
#• hist(x값, bins=x값 구간 개수, color=‘색 문자‘ 혹은 ‘색깔 문자열‘, align=‘left’/’mid’/’right’중 하나)
x= [1,1,1,2,3,2,4]
plt.hist(x, bins=4, color='green', align='mid')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y,'k--')
axs[0,1].scatter(x,y,c='black',marker='s')
axs[1,0].bar(x,y,color='red')
axs[1,1].barh(x,y,color='gray')
fig.suptitle('Drawing Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')
x = [1, 2, 3, 4]
y = [10, 4, 15, 9]
plt.plot(x,y,label='국어')
plt.title('그래프 예제')
plt.xlabel('번호')
plt.ylabel('점수')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Read the iris dataset
iris = pd.read_csv('iris.csv')

sns.scatterplot(x='petal_length',y='petal_width'
,hue='species',data=iris)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
iris=pd.read_csv('iris.csv')
sns.countplot(x='species',data=iris)
plt.show()
```

```python
iris=pd.read_csv('iris.csv')
sns.barplot(x='species',y='petal_length',data=iris, errorbar='sd')
plt.show()
```
